```python
from typing import List

class Bank:

    def __init__(self, balance: List[int]):
        balance.insert(0, None)
        self.num = len(balance)
        print(balance)
        self.balance = balance
    
    def check_account(self, *accounts):
        for account in accounts:
            if not (0 < account < self.num):
                return False
        return True

    def transfer(self, account1: int, account2: int, money: int) -> bool:
        if not self.check_account(account1, account2):
            return False

        if self.balance[account1] - money >= 0:
            self.balance[account2] += money
            self.balance[account1] -= money
            return True
        return False

    def deposit(self, account: int, money: int) -> bool:
        if not self.check_account(account):
            return False
        self.balance[account] += money
        return True

    def withdraw(self, account: int, money: int) -> bool:
        if not self.check_account(account):
            return False

        if self.balance[account] - money >= 0:
            self.balance[account] -= money
            return True
        return False



# Your Bank object will be instantiated and called as such:
# obj = Bank(balance)
# param_1 = obj.transfer(account1,account2,money)
# param_2 = obj.deposit(account,money)
# param_3 = obj.withdraw(account,money)
```

```python
import random

def replace(l, idx, new_base):
    l[idx] = new_base
    
class Gene:
    def __init__(self, dna) -> None:
        self.dna = list(dna)
        self.mutations = {
            '+': list.insert,
            '-': replace,
        }
    
    def mutation(self):
        idx = random.randint(0, len(self.dna) - 1)
        mut_type = random.choice(list(self.mutations.keys()))
        nucleotide = random.choice('ACTG')
        self.mutations[mut_type](self.dna, idx, nucleotide)
        return self, idx
    
    def break_gene(self):
        if random.choice((True, False)):
            self.dna = self.dna[:len(self.dna) // 2]
        else:
            self.dna = self.dna[len(self.dna) // 2:]
        return self

    def __repr__(self) -> str:
        return ''.join(self.dna)
```

```python
gene = Gene('ACTTGA')
```

```python
gene.mutation()
```
